```python
import numpy as np
import pandas as pd
import plotly.figure_factory as ff
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
class CLT:
  def __init__(self,data,m,n=30):
    self.data = data
    self.n = n
    self.m = m
    self.pop_mean = 0
    self.sample_means = []
    self.random_sample_mean()
    self.pop_mean = self.pop_mean / n


  def random_sample_mean(self):
    for _ in range(self.n):
      sample = np.random.choice(self.data,size=self.m)
      sample_mean = np.mean(sample)
      self.pop_mean += sample_mean
      self.sample_means.append(sample_mean)
    return self.sample_means

  def plot_clt(self,axes,i,j):
    sns.histplot(self.sample_means,ax=axes[i,j],kde=True)
```

```python

def pdf(data,step):
  pd = []
  c = 0
  l = len(data)
  s = np.min(data)
  for i in np.sort(data):
    if i <= s:
      c+=1
    else:
      pd.append(c/l)
      c=0
      s +=step
  su = sum(pd)
  pd = [i/su for i in pd]
  return pd

```

```python
import numpy as np
import plotly.express as px

# Define the two functions f(x) and g(x)
x = np.linspace(-2, 2, 400)  # Define a range of x values

def f(x):
    return np.where((x >= 0) & (x <= 1), 1, 0)

def g(x):
    return np.where((x >= 0) & (x <= 1), 1, 0)



# Perform the convolution (f * g)
convolution_result = np.convolve(f(x), g(x), mode='same')* (x[1] - x[0])  # Use the discrete convolution

# Create a figure to visualize the functions and the convolution result
fig = px.line(x=x, y=f(x), labels={'x': 'x', 'y': 'f(x)'}, title='f(x)')  # Plot f(x)
fig.add_scatter(x=x, y=g(x), mode='lines', line=dict(color='red'), name='g(x)')  # Plot g(x)
fig.add_scatter(x=x, y=convolution_result, mode='lines', line=dict(color='green'), name='f(x) * g(x)')  # Plot convolution result
fig.add_scatter(x=x, y=f(x), mode='lines', line=dict(color='blue'), name='f(x)')


# Show the plot
fig.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumtrapz

# Define a function for numerical integration using the trapezoidal rule
def trapezoidal_integration(x, y):
    integral = 0
    for i in range(1, len(x)):
        h = x[i] - x[i - 1]
        integral += (y[i] + y[i - 1]) * h / 2
    return integral

def trapezoidal_integration2(x, y):
    integral = [0]
    for i in range(1, len(x)):
        h = x[i] - x[i - 1]
        integral.append((y[i] + y[i - 1]) * h)
    return integral

# Create a sample DataFrame
data = {'x': np.linspace(0, 1, 100),
        'y': np.sin(np.linspace(0, 2 * np.pi, 100))}
df = pd.DataFrame(data)

# Compute the cumulative trapezoidal integration of the 'y' column
df['integral'] = [trapezoidal_integration(df['x'][:i+1], df['y'][:i+1]) for i in range(len(df))]

# Plot the original data and the integrated result
plt.plot(df['x'], df['y'], label='Original Data')
plt.plot(df['x'], np.cumsum(trapezoidal_integration2(df['x'],df['y'])), label='own function')
plt.plot(df['x'], cumtrapz(df['y'], df['x'], initial=0), label='own function')

plt.plot(df['x'], df['integral'], label='Cumulative Integral (Trapezoidal)', linestyle='--')
plt.title('Integration of Data Points in DataFrame (Trapezoidal Rule)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
def normilize(a,b):
  la,lb = len(a),len(b)
  n = la+lb-1
  a = np.pad(a,(0,n-la),'constant')
  b = np.pad(b,(0,n-lb),'constant')
  res = np.zeros(n)

  for i in range(n):
    for j in range(i+1):
      res[i] += a[i-j]*b[j]

  return res

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function
def f(x):
    return np.exp(-x**2)

# Generate x values
x = np.linspace(-3, 3, 1000)

# Compute the convolution of the function with its inverse
convolution_result = np.convolve(f(x), f(x)[::-1], mode='same')

# Plot the original function and the convolution result
plt.plot(x, f(x), label='Original Function')
plt.plot(x, convolution_result[:len(x)], label='Convolution Result')
plt.legend()
plt.xlabel('x')
plt.ylabel('Function Values')
plt.title('Convolution of a Function with its Inverse')
plt.show()
```
