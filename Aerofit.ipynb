```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
# import warnings
# warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/001/125/original/aerofit_treadmill.csv?1639992749')
df.head()
```

```python
a=df.groupby(['Usage','Product'],as_index=False).size()
a['sum'] =  a.groupby('Usage')['size'].transform('sum')
a['perc'] = ((a['size']/a['sum'])*100).round(2)
px.bar(a,x='Usage',y='perc',color='Product',text_auto=True)
```
