```python
import numpy as np
```

```python
a = [0,1,1,1]
ar = np.array(a).reshape((2,2))
ar
```

```python
inv = np.linalg.inv(ar)
inv
```

```python
egen = np.linalg.eig(ar)
egen
```

```python
eg_inv = np.linalg.inv(egen[1])
eg_inv
```

```python
res = np.matmul(ar,egen[1])
res
```

```python
scale = 4
s_ev = np.power(egen[0],scale)
s_ev
```

```python
resa = res[:,0]*s_ev[0]
resb = res[:,1]*s_ev[1]
print(resa,resb)
res*np.array([s_ev,s_ev])
```

```python
fin = np.array((resa,resb)).T
print(fin)
np.matmul(eg_inv,fin)
```
