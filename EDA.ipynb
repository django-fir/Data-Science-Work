```python

def fair_coin(n,b=0.5):
  e = np.random.uniform(0,1,n)
  a = map(lambda x:0 if x<=(1-b) else 1,e)
  return sum(a)/n

```

```python
si = [fair_coin(i,0.1) for i in range(1,1000)]
```

```python
px.scatter(y=si)
```

```python
def normal_form(u):
  c = np.sqrt(2*np.pi)
  ex = np.exp(-(u**2)/2)
  return ex/c
```

```python
def kde(data,h,i):
  n = len(data)
  u = np.mean(data)
  si = np.std(data)
  return np.sum(normal_form(-(data - data[i])/h))/n*h


```

```python
ar = np.random.pareto(2,1000)
pdf = [kde(ar,2,i) for i in range(len(ar))]
cdf = np.cumsum(pdf)
# print(pdf[:5],cdf[:6])
print(sum(pdf))
```

```python
px.scatter(y=pdf,x=ar)
```

```python
px.scatter(y=cdf,x=np.sort(ar))
```
