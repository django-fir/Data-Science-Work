```python
import plotly.express as px
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python


df = px.data.iris()
df.columns
df.head()
```

```python
fig = px.scatter_3d(df,x="sepal_length",y="sepal_width",z="petal_length",color="species",range_z=[0,7],color_discrete_map={"setosa":"black","versicolor":"yellow"},)
fig.show()
```

```python
df1 = px.data.carshare()
df1.head()
df1 = df1.sort_values(by="peak_hour")
```

```python
fig = px.scatter_3d(df1,x="centroid_lat",y="centroid_lon",z="car_hours",color="peak_hour",animation_frame="peak_hour")
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 100
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 500
fig.show()
```
