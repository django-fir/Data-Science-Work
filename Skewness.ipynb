```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import scipy as sp
```

```python
def skewness1(data):
  mean = np.mean(data)
  std = np.std(data)
  median = np.median(data)
  return (mean-median)/std

def skewness2(data):
  temp = ((data-np.mean(data))/np.std(data))**3
  return np.mean(temp)

```

```python
data = np.random.normal(1,3,(1000))-np.append(np.zeros((1,700)),np.random.uniform(1,10,(300)))
# data = np.random.normal(1,3,(1000))
sns.distplot(data)
```

```python
skewness1(data)
```

```python
skewness2(data)
```

```python
data = np.sort(np.random.normal(2,3,(1000)))
t = np.sort(np.random.normal(2,5,(1000)))
data = data - t
sns.distplot(data)
```

```python
def kurtosis(data):
  mean = np.mean(data)
  std = np.std(data)
  return np.mean(((data-mean)/std)**4)
```

```python
sp.stats.kurtosis(data)
```

```python
kurtosis(data)-3
```
