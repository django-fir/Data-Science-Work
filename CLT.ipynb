```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pylab
import scipy.stats as st
import plotly.graph_objects as go
```

```python
data_u = np.random.uniform(1,100,(1000))
data_n = np.random.normal(2,2,(1000))
sns.distplot(data_u)
plt.show()
sns.distplot(data_n)
plt.show()
```

```python
def clt(data,m,n):
  samples = []
  for i in range(n):
   sample = np.mean(np.random.choice(data,m))
   samples.append(sample)
  return samples

data_u = [i for i in range(0,1000)]
# clt(data_n,100,20)
```

```python
axe = [(0,0),(1,0),(2,0),(3,0),(0,1),(1,1),(2,1),(3,1)]
trail = [(100,10),(100,20),(100,40),(50,50),(500,50),(1000,50),(500,500),(500,1000),(10,200),(2,2)]
for (i,(m,n)) in enumerate(trail):
  sns.distplot(data_u)
  sns.scatterplot(x=clt(data_u,m,n),y=np.zeros(n))
  plt.title(str(trail[i]))
  plt.show()
sns.distplot(data_u)
```

```python
def qq_plot(x,y):
  xt = np.sort(x)
  yt = np.sort(y)
  return xt,yt
```

```python
x= np.random.normal(0,1,(1000))
y = np.random.normal(2,30,(1000))
x,y=qq_plot(x,y)
sns.scatterplot(x=x,y=y)
sns.lineplot(x=[x[5],x[-5]],y=[y[5],y[-5]])
plt.show()
fig = go.Figure()
fig.add_trace(go.Scatter(x=x,y=y))
fig.add_trace(go.Scatter(x=[x[5],x[-5]],y=[y[5],y[-5]]))
fig.show()
```
