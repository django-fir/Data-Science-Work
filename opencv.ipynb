```python
# TO Open Camera
camera = cv2.VideoCapture(0)
```

```python
while(camera.isOpened()):
  ret,frame = camera.read()
  cv2_imshow(frame)
  key = cv2.waitKey(0) and 0xFF
  if key == ord('q'):
    break

camera.release()
cv2.destroyAllWindows()
```
