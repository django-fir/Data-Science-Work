```python
text = """
#!/bin/bash

input_file=$1

# Set resolutions and output directory
resolutions=("640x360" "1280x720")
output_dir="output"

# Create output directory if it doesn't exist
mkdir -p "$output_dir"

# Extract video stream
ffmpeg -i "$input_file" -c:v copy -map 0:0 -f segment -segment_time 10 -segment_list "$output_dir/video.m3u8" -segment_format libx264 -segment_list_flags +live -segment_list_type m3u8 "$output_dir/video_%05d.ts"

# Extract audio streams
audio_tracks=$(ffprobe -v error -select_streams a -show_entries stream=index -of csv=p=0 "$input_file")
IFS=',' read -ra audio_array <<< "$audio_tracks"

for i in "${!audio_array[@]}"; do
  audio_index=${audio_array[$i]}
  output_audio_file="$output_dir/audio_$i.m3u8"

  ffmpeg -i "$input_file" -c:a:0 copy -map 0:"$((audio_index))" -f segment -segment_time 10 -segment_list "$output_audio_file" -segment_list_flags +live -segment_list_type m3u8 "$output_dir/audio_${i}_%05d.ts"

  # Update the master playlist with audio information
  echo "#EXT-X-MEDIA:TYPE=AUDIO,GROUP-ID='audio',LANGUAGE='eng',NAME='Audio $i',DEFAULT=$([[ $i == 0 ]] && echo YES || echo NO),AUTOSELECT=YES" >> "$output_dir/master.m3u8"
  echo "audio_$i.m3u8" >> "$output_dir/master.m3u8"
done

# Create master playlist
echo "#EXTM3U" > "$output_dir/master.m3u8"
echo "#EXT-X-STREAM-INF:BANDWIDTH=800000,RESOLUTION=${resolutions[0]}" >> "$output_dir/master.m3u8"
echo "video.m3u8" >> "$output_dir/master.m3u8"

# Serve the files from the output directory using a simple HTTP server
python -m http.server --directory "$output_dir"

"""
```

```python
with open("network.sh","w") as f:
  f.write(text)

```
