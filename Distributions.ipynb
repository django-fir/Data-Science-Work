```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
import scipy as sc
```

```python
def normal(x):
  return np.exp(-(x**2/2))/np.sqrt(2*np.pi)

def fun_exp(x):
  return np.exp(x)
```

```python
x  = np.linspace(-10,10,100)
y = normal(x)

```

```python
def integrate(x,y):
  integrals = []
  su = 0
  for i in range(1,len(x)):
    h = x[i]-x[i-1]
    res = (y[i]+y[i-1])*h/2
    su += res
    integrals.append(su)
  return integrals

def simpson_int_accurate(y,h):
  rese,reso = 0,0
  for i in range(1,len(y)-1):
    if i%2 == 0:
      rese += y[i]
    else:
      reso += y[i]
  return (h/3)*(y[0]+y[-1]+rese*2+reso*4)

def simpson_int_optimized(y,h):
  res = 0
  integrals = []
  for i in range(1,len(y)-1):
    if i%2 == 0:
      res += 2*y[i]
    else:
      res += 4*y[i]
    integrals.append(res*(h/3))
  # return (h/3)*(y[0]+y[-1]+rese*2+reso*4)
  return integrals


```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x,y=y,mode='lines',name="normal"))
fig.add_trace(go.Scatter(x=x,y=integrate(x,y),mode='lines',name="integrate"))
fig.add_trace(go.Scatter(x=x,y=sc.integrate.cumulative_trapezoid(x=x,y=y),mode='lines',name="cumulative_trapezoid"))
fig.add_trace(go.Scatter(x=x,y=[simpson_int_accurate(y[:i],x[1]-x[0]) for i in range(2,len(y)-1)],mode='lines',name="simpson_int_accurate"))
fig.add_trace(go.Scatter(x=x,y=simpson_int_optimized(y,x[1]-x[0]),mode='lines',name="simpson_int_optimized"))





fig.show()
```
