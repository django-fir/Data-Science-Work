```python
class LL:
  length = 0
  def __init__(self,d,l):
    self.data = d
    self.next = l
    self.length +=1

  def display(self):
    c = self
    while c.next != None:
      print(c.data)
      c = c.next
    print(c.data)

  def search(self,x):
    c = self
    while c.next != None and c.data != x:
      # print(self.data)
      c = c.next
    return c

  def append(self,x):
    c = self
    while c.next != None:
      c = c.next
    new = LL(x,None)
    c.next = new
    self.length +=1


  def insert(self,p,x):
    t = self.search(p)
    new = LL(x,t.next)
    t.next = new
    self.length +=1


  def delete(self,x):
    c = self
    pr = self
    while c.next!=None and c.data != x:
      pr = c
      c = c.next
      print(pr.data,c.data)
    pr.next = c.next
    self.length -=1




```

```python
class DL:
  def __init__(self,data,left,right):
    self.data = data
    self.left = left
    self.right = right

  def display(self):
    t = self
    if t == None:
      return "Empty"
    while t.right != None:
      print(t.data)
      t = t.right
    print(t.data)

  def append(self,x):
    t = self

    while  t.right != None:
      t = t.right
    t.right = DL(x,t,None)

  def insert_beging(self,x):
    self = DL(x,self,self)
    return self

  def insert_between(self,x,p):
    c = 1
    t = self
    while t.right != None and c<p:
      t = t.right
      c+=1
    t.left.right = DL(x,t.left,t.right)

  def delete(self,p):
    c = 1
    t = self
    while t.right != None and c<p:
      t = t.right
      c+=1
    t.left.right = t.right
    t.right.left = t.left


  def search(self,x):
    t = self
    p = 0
    while t.right != None:
      if t.data == x:
        return p
      p+=1




```

```python
p = 0
dl = DL(2,p,None)
dl.append(4)
dl.append(5)
dl.append(6)
dl.append(7)
dl = dl.insert_beging(1)
dl.insert_between(100,5)
dl.display()
dl.delete(3)
```
