```python
class Stack:
  def __init__(self,ele):
    self.st = ele
    self.top = len(ele)-1
    self.len = len(ele)

  def push(self,x):
    self.st.append(x)
    self.top +=1
    self.len +=1

  def pop(self):
    if not self.isempty():
      self.top -=1
      self.len-=1
      return self.st.pop()
    return "Stack Is Empty"

  def isempty(self):
    return True if self.top == -1 else False

  def stop(self):
    if self.isempty():
      return 0
    return self.st[self.top]

  def __str__(self):
    print(self.st)
    return ""


```

```python
def next_greater(x):
  s = Stack([])
  s.push(x[0])
  r = []
  for i in x[1:]:
    if s.stop() < i:
      while not s.isempty() and i > s.stop():
        s.pop()
        r.append(i)
      s.push(i)
    else:
      s.push(i)
    # print(s.st)
  for p in s.st:
    r.append(-1)
  return r


```

```python
def partion(s,e,a):
  p = a[s]
  i = s
  for j in range(s,e+1):
    if a[j] <= p:
      temp = a[j]
      a[j] = a[i]
      a[i] = temp
      i+=1
  a[s] = a[i-1]
  a[i-1] = p
  return i-1



```

```python
def quick_sort(s,e,a):
  if s < e:
    p = partion(s,e,a)
    quick_sort(s,p-1,a)
    quick_sort(p+1,e,a)
```

```python
a = [3,7,8,5,2,1,9,5,4]
a = [11,11,-1,5,6,4,2,8,10,]

quick_sort(0,len(a)-1,a)
print(a)
```

```python
def selection_sort(a,l):
  p = 0
  for i in range(l-1):
    max = 0
    for j in range(l-i):
      if a[max]<a[j]:
        max = j
      p = j
    if max != p: #inovative: (think once) reduces no(swaps) to zero
      temp = a[p]
      a[p] = a[max]
      a[max] = temp
```

```python
a = [9,2,2,8,2,7,6,1,1,1,2,4,3,2,1,2,5,5]
selection_sort(a,len(a))
print(a)
```

```python
def counting_sort1(a,l):
  c = [0 for i in range(max(a)+1)]
  for i in a:
    c[i] +=1
  i = 0
  j = 0
  while l > j:
    if c[i] != 0:
      a[j] = i
      c[i]-=1
      j+=1
    else:
      i+=1




```

```python
def counting_sort2(a,l):
  l = max(a)
  c = [0 for i in range(l+1)]
  for i in a:
    c[i] +=1
  i = 0
  j = 0
  while l >= i:
    if c[i] != 0:
      a[j] = i
      c[i]-=1
      j+=1
    else:
      i+=1
```

```python
a = [6,6,3,3,2,2,10,10,3,3,6,1,1]

counting_sort2(a,len(a))
print(a)
```

```python
def insertion_sort(a,l):
  for i in range(1,l):
    temp = a[i]
    j = i-1
    while temp < a[j] and j>=0:
      a[j+1] = a[j]
      j-=1
    a[j+1] = temp
```

```python
a = [5,4,6,3,8,2,10,9,5,5]

insertion_sort(a,len(a))
print(a)
```

```python
def mearge(A,s,m,e):
  n1 = m-s+1
  n2 = e-m
  k = s
  a = A[s:m+1]
  b = A[m+1:e+1]
  i = 0
  j = 0
  while i<n1 and j<n2:
    if a[i] <= b[j]:
      A[k] = a[i]
      i+=1
    else:
      A[k] = b[j]
      j+=1
    k+=1
  while i<n1:
    A[k] = a[i]
    k+=1
    i+=1
  while j<n2:
    A[k] = b[j]
    k+=1
    j+=1


```

```python
def mearge_sort(A,s,e):
  if s<e:
    mid = (s+e)//2
    mearge_sort(A,s,mid)
    mearge_sort(A,mid+1,e)
    mearge(A,s,mid,e)
```

```python
a = [5,4,6,3,8,2,10,9,5,5]
mearge_sort(a,0,len(a)-1)
print(a)
```
