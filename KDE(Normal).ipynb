```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

##Gausian Distribution(finding the distribution) from scrach

```python
#create gausian distribution
def normal_distribution():
  values = np.random.normal(4,5,(1000))
  return values
```

```python
def normal(u):
  c = 1/np.sqrt(2*np.pi)
  return np.exp(-(u)**2/2)*c

def optimum_h(std,n):
  return np.power(4*(std)**5/(3*n),1/5)
def pdf_kde(data,h):
  n = len(data)
  pdfs = []
  for i in data:
    temp = np.sum(normal((data-i)/h))/(n*h)
    pdfs.append(temp)
  return pdfs


data = np.sort(normal_distribution())
h = optimum_h(np.std(data)+0.5,len(data))
print(h)
pdf = pdf_kde(data,h)
print(np.sum(pdf))
```

```python
px.scatter(x=data,y=pdf)
```

```python
sns.distplot(data)
plt.show()
```

```python
def intigration(x,y):
  integrls = []
  for i in range(1,len(x)):
    dif = ((x[i]-x[i-1])*(y[i]+y[i-1]))/2
    integrls.append(dif)
  return integrls

intef = intigration(data,pdf)
cdf = np.cumsum(intef)
total = np.sum(intef)
std = np.std(data)
mean = np.mean(data)


def filter_data(std,m,data,c1=1,c2=1):
  datas = []
  pdfs = []
  for i in range(len(data)):
    if data[i] < c1*std+m and data[i]>m+std*c2:
      datas.append(data[i])
      pdfs.append(pdf[i])
  return (datas,pdfs)

d,p = filter_data(std,mean,data,c1=2,c2=-2)
inter = sum(intigration(d,p))
print(inter/total)



px.scatter(x=data[1:],y=cdf)



```

```python
def pdf_cdf(data,cdf):
  pdfs1 = []
  for i in range(1,len(data)):
    dx = data[i]-data[i-1]
    dy = cdf[i]-cdf[i-1]
    pdfs1.append(dy/dx)
  return pdfs1

yy = pdf_cdf(data[1:],cdf)
print(yy[:2])
px.scatter(x=data[1:],y=yy+[0])
```

```python
data = [np.random.lognormal(1,0.5,1000),np.random.lognormal(2,0.25,1000),np.random.lognormal(3,1,1000)]


def dist(data):
  for i in data:
    sns.distplot(i)
    print(np.mean(i),np.log(np.mean(i)),sep="\t")
  plt.show()


dist(data)
```

```python
def log_no(data):
  for i in data:
    d = np.log(i)
    print(np.mean(d),np.exp(np.mean(d)),np.std(d),sep="\t")
    sns.distplot(d)
  plt.show()

log_no(data)
```
