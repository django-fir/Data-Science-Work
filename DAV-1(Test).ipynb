```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the two PDFs
def f1(t):
    return np.exp(-t) * (t >= 0)

def f2(t):
    return 2 * np.exp(-2 * t) * (t >= 0)

# Define the convolution function
def convolve(f1, f2, t):
    result, _ = quad(lambda x: f1(x) * f2(t - x), 0, t)
    return result

# Generate the time values
t_values = np.linspace(0, 5, 400)

# Compute the convolution values
convolution_values = [convolve(f1, f2, t) for t in t_values]

# Plot the original PDFs and their convolution
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(t_values, [f1(t) for t in t_values], label='$f_1(t)$')
plt.title('PDF of Module 1')
plt.xlabel('Time (t)')
plt.ylabel('Probability Density')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(t_values, [f2(t) for t in t_values], label='$f_2(t)$', color='orange')
plt.title('PDF of Module 2')
plt.xlabel('Time (t)')
plt.ylabel('Probability Density')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(t_values, convolution_values, label='$f_1 * f_2(t)$', color='green')
plt.title('Convolution (Total Execution Time)')
plt.xlabel('Time (t)')
plt.ylabel('Probability Density')
plt.legend()

plt.tight_layout()
plt.show()

```

```python
# Correcting the padding and visualization

# Define the two discrete functions
f = np.array([2, 1, 0, -1])  # Function f[k]
g = np.array([1, 0.5, 0])    # Function g[k]

# Prepare for convolution visualization
f_length = len(f)
g_length = len(g)
conv_length = f_length + g_length - 1

# Create padded versions of g for visualization
h = np.zeros(conv_length)  # Convolution result

# Visualize step-by-step convolution
fig, axs = plt.subplots(conv_length, 1, figsize=(8, conv_length * 2))
fig.tight_layout(pad=4.0)

# Perform the convolution manually with visualization
for n in range(conv_length):
    # Calculate overlap range
    overlap_start = max(0, n - (g_length - 1))
    overlap_end = min(f_length, n + 1)

    # Create shifted version of g
    shifted_g = np.zeros(f_length)
    for i in range(overlap_start, overlap_end):
        shifted_g[i] = g[n - i]

    # Multiply overlapping values
    products = f * shifted_g
    # Compute the sum of products
    h[n] = np.sum(products)

    # Plot the current state of f and shifted g
    axs[n].stem(range(f_length), f, linefmt='C0-', markerfmt='C0o', basefmt='C0-', label='f[k]')
    axs[n].stem(range(f_length), shifted_g, linefmt='C1--', markerfmt='C1x', basefmt='C1-', label='g[n-k]')
    axs[n].stem(range(f_length), products, linefmt='C2:', markerfmt='C2s', basefmt='C2-', label='f[k] * g[n-k]')

    # Annotate the sum of products
    axs[n].set_title(f'Step {n+1}: h[{n}] = Sum(f[k] * g[n-k]) = {h[n]:.2f}')
    axs[n].legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the functions f(τ) = τ^2 and g(τ) = τ
def f(tau):
    return tau**2

def g(tau):
    return tau

# Define the convolution integral h(t)
def h(t):
    # Integrand: f(τ) * g(t - τ)
    integrand = lambda tau: f(tau) * g(t - tau)
    # Integrate over the range [0, t]
    result, _ = quad(integrand, 0, t)
    return result

# Generate values for t to compute h(t)
t_values = np.linspace(0, 5, 100)  # t from 0 to 5
h_values = [h(t) for t in t_values]

# Plot the functions f(τ), g(t-τ), and the result h(t)
fig, ax = plt.subplots(3, 1, figsize=(10, 12))
tau_values = np.linspace(0, 5, 100)

# Plot f(τ)
ax[0].plot(tau_values, f(tau_values), label=r"$f(\tau) = \tau^2$", color="blue")
ax[0].set_title("Function $f(\\tau)$")
ax[0].set_xlabel(r"$\tau$")
ax[0].set_ylabel(r"$f(\tau)$")
ax[0].legend()
ax[0].grid(True)

# Plot g(t-τ) for a specific t
t_example = 3  # Example t to visualize
g_shifted = [g(t_example - tau) for tau in tau_values]
ax[1].plot(tau_values, g_shifted, label=fr"$g(t - \tau)$, $t={t_example}$", color="orange")
ax[1].set_title(f"Shifted Function $g(t - \\tau)$ for $t={t_example}$")
ax[1].set_xlabel(r"$\tau$")
ax[1].set_ylabel(r"$g(t-\tau)$")
ax[1].legend()
ax[1].grid(True)

# Plot the convolution result h(t)
ax[2].plot(t_values, h_values, label=r"$h(t) = \int f(\tau) g(t - \tau) d\tau$", color="green")
ax[2].set_title("Convolution Result $h(t)$")
ax[2].set_xlabel(r"$t$")
ax[2].set_ylabel(r"$h(t)$")
ax[2].legend()
ax[2].grid(True)

plt.tight_layout()
plt.show()
```
