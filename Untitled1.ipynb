```python
import numpy as np
from datetime import datetime
```

```python
a = [1,2,3,4]
b = [5,6,7,8]
```

```python
def dot_prod(a,b):
  result = 0
  for i,j in zip(a,b):
    result += i*j
  return result
```

```python
before = datetime.now()
for _ in range(100000):
  dot_prod(a,b)
after = datetime.now()
print(after-before)
```

```python
before1 = datetime.now()
for _ in range(100000):
  np.dot(a,b)
after1 = datetime.now()
print(after1-before1)
```
