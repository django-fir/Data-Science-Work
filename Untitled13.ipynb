```python
def find_missing_xor(a,n):
  total = 1
  for i in range(2,n+1):
    total = total ^ i
  for j in a:
    total = total ^ j
  return total
```

```python
import random
```

```python
a = [1,2,3,4,6,7,8,9,10]
random.shuffle(a)
n = 10
find_missing_xor(a,n)
```
