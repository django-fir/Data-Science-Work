```python
import requests

bot_token = token
channel_username = "@"
message_id = 1532145811  # Replace with the actual message ID

# Get information about the message
url_message = f"https://api.telegram.org/bot{bot_token}/getChat?chat_id={channel_username}&message_id={message_id}"
response_message = requests.get(url_message)
message_info = response_message.json()['result']
print(response_message.json())

# Extract files from message_info as needed
files = message_info.get('photo', []) + message_info.get('document', []) + message_info.get('video', [])
print(files)
# Process the list of files
for file_info in files:
    file_id = file_info['file_id']

    # Use the file_id to get file information
    url_file = f"https://api.telegram.org/bot{bot_token}/getFile?file_id={file_id}"
    response_file = requests.get(url_file)
    file_path = response_file.json()['result']['file_path']

    # Download the file
    file_download_url = f"https://api.telegram.org/file/bot{bot_token}/{file_path}"
    downloaded_file = requests.get(file_download_url)

    # Save the file locally or process it as needed
    with open(f"downloaded_file_{file_id}.jpg", "wb") as f:
        f.write(downloaded_file.content)
```
