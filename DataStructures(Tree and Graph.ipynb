```python
#Node
class Node:
  def __init__(self,data):
    self.data = data
    self.left = None
    self.right = None

  def __str__(self):
    return f"{self.data}"

```

```python
def display_tree(root, level=0):
  # Base Case
  if root is None:
    return

  # Print right subtree with proper indentation
  display_tree(root.right, level + 1)

  # Print current node data with spacing
  print(" " * (level * 4) + str(root.data))

  # Print left subtree with proper indentation
  display_tree(root.left, level + 1)
```

```python
def posi(mat,b):
  v = [1,2,3,4,5,6,7,8,9]
  r,c = b
  for i in range(len(mat)):
    el = mat[r][i]
    if el in v:
      v.remove(el)
      # print("col:",el)



  for j in range(len(mat)):
    if mat[j][c] in v:
      v.remove(mat[j][c])
      # print("row:",mat[j][c])

  ri = (r//3)*3
  cj = (c//3)*3
  # print(ri,cj)
  for i in range(ri,ri+3):
    for j in range(cj,cj+3):
      if mat[i][j] in v:
        v.remove(mat[i][j])
        # print("sqare:",mat[i][j])


  # print("position:",v,ri,cj,r,c)
  return v



```

```python
def blanks(mat,n):
  b = []
  for i in range(n):
    for j in range(n):
      if mat[i][j] == 0:
        b.append((i,j))
  return b
```

```python
def suduku(mat,b,n,bl):
  if b==n:
      # print(mat)
      return mat
  r,c = bl[b]
  for i in posi(mat,bl[b]):
    mat[r][c] = i
    # print(mat,r,c,i)
    if valid_mat(mat,bl[b]):
      a= suduku(mat,b+1,n,bl)
      if a:
        return a
      mat[r][c] = 0




```

```python
def attack(mat,q,n):
  r,c = q
  for i in range(n+1):
    if i!= r and mat[i][c] == 1:
      return 0
    if i!=c and mat[r][i] == 1:
      return 0
  tr,tc=r,c
  while tr-1>=0 and tc-1>=0:
    tr-=1
    tc-=1
    if mat[tr][tc]==1:
      return 0
  tr,tc=r,c
  while tr-1>=0 and tc+1<=n:
    tr-=1
    tc+=1
    if mat[tr][tc]==1:
      return 0
  tr,tc=r,c
  while tr+1<=n and tc-1>=0:
    tr+=1
    tc-=1
    if mat[tr][tc]==1:
      return 0
  tr,tc=r,c
  while tr+1<=n and tc+1<=n:
    tr+=1
    tc+=1
    if mat[tr][tc]==1:
      return 0
  return 1



```

```python
def n_queens(mat,q,n):
  r,c = q
  if r==n+1:
    # print("Result:-",mat)
    return mat
  for i in range(n+1):
    mat[r][i] = 1
    q = (r,i)
    # print(mat,q)
    if attack(mat,q,n):
      q = (r+1,0)
      a = n_queens(mat,q,n)
      if a:
        return a
      mat[r][i] = 0

    else:
      mat[r][i] = 0
      # return a




```
