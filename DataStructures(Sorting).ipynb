```python
def insertion_sort(arr,n):
  for i in range(1,n):
    j = i-1
    pi = arr[i]
    while j>=0 and pi<arr[j]:
      arr[j+1] = arr[j]
      j-=1
      # print(arr[j+1],arr[j])
    arr[j+1] = pi
```

```python
#Bubble Sort
def bubble_sort(arr,n):
  for i in range(n-1):
    swaps = 0
    for j in range(n-i-1):
      if arr[j]> arr[j+1]:
        temp = arr[j]
        arr[j] = arr[j+1]
        arr[j+1] = temp
        swaps+=1
    if swaps<1:
      return 0



    print(arr)
```

```python
#Selection Sort
def selection_sort(arr,n):
  for i in range(n-1):
    max = 0
    for j in range(n-i):
      if arr[max]<arr[j]:
        max = j
    if max != j:
      temp = arr[max]
      arr[max] = arr[j]
      arr[j] = temp

```

```python
#Counting Sort
def counting_sort(arr,b):
  n = max(arr)+1
  a = [0 for i in range(n)]
  for i in arr:
    a[i] +=1

  for i in range(1,n):
    a[i] +=a[i-1]
  for j in arr[::-1]:
    b[a[j]-1] = j
    a[j]-=1
  return b



```

```python
#Doubly Linked List
class DLL:
  def __init__(self,data):
    self.prev = None
    self.next = None
    self.data = data
  def __str__(self):
    return f'{self.data}-->'
```

```python
def append(head,data):
  a = head
  while a.next:
    a = a.next
  d = DLL(data)
  a.next = d


def search(head,data):
  a = head
  while a.next:
    if a.data == data:
      return a
    a = a.next


def pop(head,data):
  e = search(head,data)
  e.prev.next = e.next
  e.next.prev = e.prev
  print(e)

def insert(head,data,newdata):
  e = search(head,data)
  d = DLL(newdata)
  d.prev = e
  d.next = e.next
  e.next.prev = d
  e.next = d


```

```python
#Stack
class Stack:
  def __init__(self):
    self.list = []
    self.top = -1
  def push(self,ele):
    self.top+=1
    self.list.append(ele)
    print(f"Pushed To Stack {ele}/{self.top}")

  def is_empty(self):
    return 0 if self.top > -1 else 1

  def pop(self):
    if not self.is_empty():
      e = self.list[self.top]
      self.list.pop()
      self.top -=1
      return e
    print("Empty")

  def stop(self):
    if not self.is_empty():
      return self.list[self.top]
    return -1

  def __str__(self):
    temp = self.top
    while temp >-1:
      print(self.list[temp])
      temp-=1
    return ''

```

```python
#Parentheses Matching
def convert(s):
  d = {
      "{":3,
      "}":-3,
      "(":1,
      ")":-1,
      "[":2,
      "]":-2,
  }
  return d[s]
def para_match(s):
  st = Stack()
  # st.push(s[0])
  for i in s:
    ci  = convert(i)
    if ci>0:
      st.push(ci)
      print(i,ci)
    elif st.stop()+ci == 0:
      st.pop()
      print(i,ci)
    else:
      break
  if st.is_empty():
    print("Valid")
  else:
    print("Invalid")
```

```python
#Other implimetation
def priority(x):
  if x == "(":
    return 0
  elif x == "+" or x == "-":
    return 1
  elif x == "*" or x == "/":
    return 2
```

```python
def infix_to_postfix(x):
  st = Stack()
  f = ''
  for i in x:
    if i.isnumeric():
      f+=i
      print(i)
    elif i == "(":
      st.push(i)
    elif i == ")":
      x = st.pop()
      while x != '(':
        f+=x
        x = st.pop()
    elif st.is_empty():
      st.push(i)
    else:
      while priority(st.stop()) >= priority(i):
        f += st.pop()
        if st.is_empty():
          break
      st.push(i)
  while not st.is_empty():
    f+=st.pop()
  return f

```

```python
def infix_prepix(x):
  f = ''
  for i in x:
    if x=="(":
      f+=')'
    elif x==")":
      f+='('
    else:
      f+=i
  s = infix_to_postfix(x)
  return s[::-1]
```

```python
def evaluate_postfix(x):
  st = Stack()
  for i in x:
    if i.isnumeric():
      st.push(i)
    else:
      e2 = st.pop()
      e1 = st.pop()
      g = f'{int(e1)}{i}{int(e2)}'
      # print(f"exp: {g}")
      f = eval(g)
      st.push(f)
  return st.stop()
```

```python
#Quees
class Queues:
  def __init__(self,n):
    self.list = [0 for i in range(n)]
    self.top = -1
    self.bottom = 0
    self.len = n-1
    self.num = 0

  def enqueue(self,data):
    if self.top >= self.len:
      if self.num >= self.len:
        print("Full")
      else:
        print("came")
        self.top = -1
    print(self.top)
    self.top+=1
    self.list[self.top] = data
    self.num +=1

  def dequeue(self):
    if self.num >= 0:
      ele = self.list[self.bottom]
      self.bottom +=1
      self.num-=1
      if self.bottom >= self.len:
        self.bottom =0
    return ele



```

```python
#Symentric Matrix
class Matrix:
  def __init__(self,n,m):
    self.mat = []
    self.n = n
    self.m = m

  def convert(self,mat):
    for i in range(self.n):
      for j in range(i+1):
        self.mat.append(mat[i][j])

  def __len__(self):
    return len(self.mat)

  def __str__(self):
    for i in range(self.n):
      for j in range(self.m):
        print(f"{mat[i,j]},",end=" ")
      print()
    return ''



  def __getitem__(self,index):
    i,j = index
    if i < j:
      i = index[1]
      j = index[0]
    i = (i*(i+1)//2)+j
    return self.mat[i]
```
