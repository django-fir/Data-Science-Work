```python
df = pd.read_csv('uber-data.csv', dayfirst=True, na_values="NA")
df.head(10)
```

```python
df['Status'].value_counts()
```

```python
df[df['Status']=='Trip Completed']['Request timestamp'].isnull().sum()
```

```python
df['Request_parced_time']=pd.to_datetime(df['Request timestamp'],format='%d/%m/%Y %H:%M',errors='coerce')
df.loc[df['Request_parced_time'].isna(),'Request_parced_time'] = pd.to_datetime(df['Request timestamp'][df['Request_parced_time'].isna()],format='%d-%m-%Y %H:%M:%S',errors='coerce')
```

```python
df['Drop_parced_time']=pd.to_datetime(df['Drop timestamp'],format='%d/%m/%Y %H:%M',errors='coerce')
df.loc[df['Drop_parced_time'].isna(),'Drop_parced_time'] = pd.to_datetime(df['Drop timestamp'][df['Drop_parced_time'].isna()],format='%d-%m-%Y %H:%M:%S',errors='coerce')
```

```python
df['week_number'] = df['Request_parced_time'].dt.isocalendar().week    # Week number in ISO format
df['quarter'] = df['Request_parced_time'].dt.quarter                   # Quarter of the year
df['day_of_year'] = df['Request_parced_time'].dt.dayofyear
df['month'] = df['Request_parced_time'].dt.month
df['hour'] = df['Request_parced_time'].dt.hour
```

```python
df.head()
```

```python
df_grouped = df.groupby(['hour','Status'])['week_number'].count().reset_index()
```

```python
df_sum = df_grouped.groupby('hour')['week_number'].transform(lambda x:x.sum())
```

```python
df_grouped['per']=df_grouped['week_number']/df_sum*100
```

```python
df_grouped
```

```python
px.line(df_grouped,x='hour',y='per',color='Status')
```

```python
px.bar(df_grouped,x='hour',y='per',color='Status')
```

```python
tran = df[df['Status']=='Trip Completed']
tran.head()
```

```python
tran['difference']=(df['Drop_parced_time']-df['Request_parced_time']).dt.total_seconds()/60
tran.head()
```

```python
cta_time = tran.groupby('hour')['difference'].mean().reset_index()
```

```python
# atc_time = tran[tran['Pickup point']==''].groupby('hour')['difference'].mean().reset_index()
```

```python
px.line(cta_time,x='hour',y='difference')
```

```python
px.histogram(tran,'difference',color='Pickup point')
```

```python
# using cut method
times = [0, 4, 9, 16, 21, 24]
values = ['Dawn', 'Early Morning', 'Noon', 'Late Evening ', 'Night']
df['time']=pd.cut(x=df['hour'], bins=times, labels=values)
```
