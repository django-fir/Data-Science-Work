```python
class Stack:
  def __init__(self):
    self.stk = []
    self.top = -1

  def push(self,x):
    self.stk.append(x)
    self.top += 1

  def pop(self):
    resu = self.stk.pop() if self.top != -1 else "Empty"
    self.top -=1
    return resu
  def topi(self):
    return self.stk[self.top] if self.top != -1 else "Empty"

  def isempty(self):
    return True if self.top == -1 else False

  def display(self):
    print("___")
    for i in self.stk[::-1]:
      print(i)
    print("^^^")

    if self.top == -1:
      print("Empty")

```

```python
s = Stack()
s.push(1)
s.push(2)
s.push(10)
print(s.pop())
s.display()
s.topi()
```

```python
op_to_n = {
    '+':1,
    '-':1,
    '*':2,
    '/':3,
    '(':9,
    ')':8,
    '^':4,
}
n_to_op = {
    1:'+',
    1:'-',
    2:'*',
    3:'/',
    9:'(',
    0:')',
    4:'^',
}
```

```python
def infix_to_postfix(exp):
  res = ''
  s = Stack()
  for i in exp:
    if i.isdigit():
      res+=i
    else:
      nu = op_to_n[i]
      if s.isempty():
        s.push(i)
      elif op_to_n[s.topi()] > nu:
        if s.topi() == "(":
          s.push(i)
        else:
          t = s.pop()
          res+=t
          s.push(i)
      elif op_to_n[s.topi()] == nu:
        res += s.pop()
        s.push(i)
      elif op_to_n[s.topi()] < nu:
        if i == ")":
          while s.topi() != "(":
            res+= s.pop()
          s.pop()
        else:
          s.push(i)
      # s.display()
  while not s.isempty():
    res += s.pop()


  return res



```

```python
infix_to_postfix('(1+2*3)*3/4^5')
```
