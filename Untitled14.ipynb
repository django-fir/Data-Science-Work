```python
def return_value(contact_box,att):
  ob = None
  try:
    ob = contact_box.find(class_=att)
    if ob:
      if att == "getdirectin":
        ob = contact_box.findAll(class_=att)
        # print(ob)
        return ob[1].get('href')
      return ob.get_text(strip=True)
    else:
      return ''
  except:
    print(ob)
```

```python


# Extract text inside each class
def get_shops(soup):
  all = []
  contact_boxes = soup.find_all('div', class_='contact-boxright mbottom delivery-rightgreybox')
  for contact_box in contact_boxes:
      result = {}
      title = return_value(contact_box,'right-titlebox')
      address = return_value(contact_box,'black-color')
      mobile_number = return_value(contact_box,'fw-400')
      url = return_value(contact_box,'getdirectin')
      lat = return_value(contact_box,'lat')
      lan = return_value(contact_box,'lng')
      result["title"]= title
      result["address"]= address
      result["mobile"]= mobile_number
      result["url"]=url
      result["lat"]=lat
      result["lan"]=lan
      # print("DONE")
      all.append(result)
  return all
```

```python
shops = []
for i in result:
  soup = BeautifulSoup(i.replace('\\"','"'), 'html.parser')
  shops.append(get_shops(soup))

print(len(shops),shops)
```

```python
tr = 1
for i in shops:
  for j in i:
    j['dist_id'] = tr
    j['sta_id'] = 2
  tr+=1
```
