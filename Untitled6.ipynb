```python
class Solution(object):
    def wordSubsets(self, words1, words2):
        """
        :type words1: List[str]
        :type words2: List[str]
        :rtype: List[str]
        """

        set_word2 = set(words2)
        result = []
        for i in words1:
            if set_word2.issubset(set(i)):
                result.append(i)
        return result

s = Solution()
p = s.wordSubsets(["amazon","apple","facebook","google","leetcode"]
,["l","e"])
print(p)
```
