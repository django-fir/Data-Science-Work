```python
class Graph:
  def __init__(self,mat):
    self.mat = mat
    self.n = len(mat)
    self.d = {}
    self.f = {}
    self.t = 0
    self.comp = []
    self.conn = []
    self.colour = {i:'W' for i in range(self.n)}
    self.edges = {}
    self.sets = []
    self.set_index = {}

  def bfs(self,source):
    colour = {i:'W' for i in range(self.n)}
    parents = {0:0}
    Q = deque()
    colour[source]='G'
    Q.append(source)
    while len(Q) != 0:
      node = Q.popleft()
      colour[node] = 'B'
      for i in range(self.n):
        if self.mat[node][i] != 0 and colour[i] == 'W':
          colour[i] = 'G'
          Q.append(i)
          parents[i] = parents[node]+1




    return parents

  def dfs(self,source):
    self.t=0
    self.d = {}
    self.f = {}
    self.conn,self.comp = [],[]
    self.colour = {i:'W' for i in range(self.n)}
    # ------connected components---|^|^|^|
    self.colour[source] = 'G'
    self.d[source] = 0
    self.dfs_r(source,self.mat)
    for i in range(self.n):
      if self.colour[i] == 'W':
        self.dfs_r(i,self.mat)
    return self.d,self.f


  def dfs_r(self,i,mat):
    # print(i)
    self.colour[i] = 'G'
    self.t+=1
    self.d[i]=self.t
    self.conn.append(i)
    for j in range(self.n):
      if mat[i][j]!=0 and self.colour[j] == 'W':
        self.colour[j] = 'G'
        # self.d[j] = self.t
        # self.t+=1
        self.dfs_r(j,mat)
    self.colour[i] = 'B'
    self.f[i] = self.t+1
    self.t+=1
    self.comp.append(self.conn)
    self.conn = []

  def transpose(self):
    mat_t = []
    for i in range(self.n):
      temp = []
      for j in range(self.n):
        temp.append(self.mat[j][i])
      mat_t.append(temp)
    return mat_t



  def connected_components(self,source):
    d,f = self.dfs(source)
    print(d,f)
    f_inv = {v:k for k,v in f.items()}
    sorted_f = sorted(f,key=lambda x:f[x],reverse=True)
    print(sorted_f)
    self.t=0
    self.d = {}
    self.f = {}
    self.conn,self.comp = [],[]
    self.colour = {i:'W' for i in range(self.n)}
    mat_t = self.transpose()
    for i in sorted_f:
      if self.colour[i] == 'W':
        self.dfs_r(i,mat_t)
    return self.comp

  def compute_edges(self):
    for i in range(self.n):
      self.set_index[i]=i
      self.sets.append(set([i]))
      for j in range(i+1,self.n):
        w = self.mat[i][j]
        if w != 0:
          self.edges[f'{i}-{j}'] = w
    return self.edges

  def krus_kal(self):
    A = []
    sort_edge = sorted(self.edges,key=lambda x:self.edges[x])
    # print(sort_edge)
    for i in sort_edge:
      f,s = i.split('-')
      fc,sc = int(f),int(s)
      f = self.set_index[fc]
      s = self.set_index[sc]
      if f!=s:
        A.append((fc,sc))
        self.sets[s]=self.sets[f].union(self.sets[s])
        for j in self.sets[f]:
          self.set_index[j]=s
        self.sets[f] = set()
        print((fc,sc),(f,s),self.set_index,self.sets)

    return A,self.sets


















```
